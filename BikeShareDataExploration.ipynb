```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import csv
import time
import scipy.stats as st
from scipy.stats import linregress
```

```python
## Output File (CSV)
data_file_hour="archive/hour.csv"
data_file_day="archive/day.csv"
#Creating DataFrame
Bike_Share_Day_df=pd.read_csv(data_file_day)
Bike_Share_Hour=pd.read_csv(data_file_hour)
Bike_Share_Hour.head()
```

```python
Bike_Share=Bike_Share_Hour[['instant','dteday','season','yr','mnth','hr',
                                'holiday','weekday','workingday','weathersit','cnt' ]].copy()
Bike_Share.head()
```

# Adding Season labels

```python
Bike_Share_Season=Bike_Share.copy()
bins=[0,1,2,3,4]
group=['Spring','Summer','Fall','Winter']
Bike_Share_Season['Season Name']=pd.cut(Bike_Share_Season['season'],bins,labels=group,include_lowest=True)
Bike_Share_Season.head()
```

# Adding Weather labels

```python
#weathersit : 
#		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
#		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
#		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
#		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
```

```python
Bike_Share_Day=Bike_Share_Day_df[['instant','dteday','season','yr','mnth',
                                'holiday','weekday','workingday','weathersit','cnt' ]].copy()
bins=[0,1,2,3,4]
group=['Clear','Cloudy','Light Snow/Rain','Heavy Rain']
Bike_Share_Day['Weather Name']=pd.cut(Bike_Share_Day['weathersit'],bins,labels=group,include_lowest=True)
Bike_Share_Day.head()
```

```python
#Output to csv file
Bike_Share_Day.to_csv('Resources/Bike_Share_Day.csv', index=False ,header=True)
Bike_Share_Season.to_csv('Resources/Bike_Share_Season.csv', index=False ,header=True)
```
